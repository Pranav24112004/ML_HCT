```python
def categorize_age(age):
    if age <= 12:
        return '(0-12)'
    elif 13 <= age <= 17:
        return '(13-17)'
    elif 18 <= age <= 65:
        return '(18-65)'
    else:
        return '(65+)'

# Apply the function to categorize both donor_age and age_at_hct
df['donor_age'] = df['donor_age'].apply(categorize_age)
df['age_at_hct'] = df['age_at_hct'].apply(categorize_age)

pd.set_option('display.max_rows',None)
df.head(2)
```

```python
for col in df1.columns:
    unique_values = df1[col].unique()
    print(f"{col}: {unique_values}")
```

```python
from sklearn.feature_selection import f_regression, SelectKBest, mutual_info_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LassoCV
X_encoded = df1.drop(columns=["efs_time"])  # Encoded categorical features
y = df1["efs_time"]  # Continuous target variable

# Compute ANOVA F-statistics
f_values, p_values = f_regression(X_encoded, y)

# Select top 50 features (initial filtering)
top_k = 50
selected_filter = SelectKBest(score_func=f_regression, k=top_k)
X_filtered = selected_filter.fit_transform(X_encoded, y)

# Get selected feature names
selected_filter_features = X_encoded.columns[selected_filter.get_support()]
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_filtered, y)

# Get feature importance
importance_df = pd.DataFrame({
    'Feature': selected_filter_features,
    'Importance': rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Keep only top 30 features based on importance
top_rf_features = importance_df.head(30)['Feature'].values
X_rf_selected = X_encoded[top_rf_features]
lasso = LassoCV(cv=5)  # Cross-validation to find best alpha
lasso.fit(X_filtered, y)

# Get non-zero coefficient features
selected_lasso_features = selected_filter_features[lasso.coef_ != 0]
X_lasso_selected = X_encoded[selected_lasso_features]
# Compute Mutual Information scores
mi_scores = mutual_info_regression(X_filtered, y)  # Uses only the 50 pre-selected features

mi_importance_df = pd.DataFrame({
    'Feature': selected_filter_features,  # Use filtered features instead of full dataset
    'MI_Score': mi_scores
}).sort_values(by='MI_Score', ascending=False)

# Select top 30 features from Mutual Information (within the 50 features)
top_mi_features = mi_importance_df.head(30)['Feature'].values

# Select top 30 features based on Mutual Information
top_mi_features = mi_importance_df.head(30)['Feature'].values
X_mi_selected = X_encoded[top_mi_features]
# Combine all selected features from different methods
final_features = set(top_rf_features) | set(selected_lasso_features) | set(top_mi_features)
X_final_selected = X_encoded[list(final_features)]

print(f"Final selected features count: {len(final_features)}")
```
